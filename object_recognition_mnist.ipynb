```python

# defining all the prerequisite functions as asked in the assignment
def calculate_accuracy(y_true, y_pred):
    correct_predictions = np.sum(y_true == y_pred)
    total_predictions = len(y_true)
    accuracy = correct_predictions / total_predictions
    return accuracy

def calculate_balanced_accuracy(y_true, y_pred):
    true_positive = np.sum((y_true == 1) & (y_pred == 1))
    true_negative = np.sum((y_true == -1) & (y_pred == -1))
    false_positive = np.sum((y_true == -1) & (y_pred == 1))
    false_negative = np.sum((y_true == 1) & (y_pred == -1))

    sensitivity = true_positive / (true_positive + false_negative)
    specificity = true_negative / (true_negative + false_positive)

    balanced_accuracy = (sensitivity + specificity) / 2
    return balanced_accuracy

def calculate_auc_roc(y_true, y_prob):
    fpr, tpr, threshold = roc_curve(y_true, y_prob)
    auc_value = auc(fpr, tpr)
    return auc_value

def calculate_auc_pr(y_true, y_prob):
    precision, recall, _ = precision_recall_curve(y_true, y_prob)
    auc_value = auc(recall, precision)
    return auc_value

def calculate_f1_score(y_true, y_pred):
    true_positive = np.sum((y_true == 1) & (y_pred == 1))
    false_positive = np.sum((y_true == -1) & (y_pred == 1))
    false_negative = np.sum((y_true == 1) & (y_pred == -1))

    precision = true_positive / (true_positive + false_positive)
    recall = true_positive / (true_positive + false_negative)

    f1 = 2 * (precision * recall) / (precision + recall)
    return f1

def calculate_matthews_corrcoef(y_true, y_pred):
    true_positive = np.sum((y_true == 1) & (y_pred == 1))
    true_negative = np.sum((y_true == -1) & (y_pred == -1))
    false_positive = np.sum((y_true == -1) & (y_pred == 1))
    false_negative = np.sum((y_true == 1) & (y_pred == -1))

    numerator = (true_positive * true_negative) - (false_positive * false_negative)
    denominator = np.sqrt((true_positive + false_positive) * (true_positive + false_negative) *
                         (true_negative + false_positive) * (true_negative + false_negative))

    mcc = numerator / denominator if denominator != 0 else 0
    return mcc



```

```python
#defining metric functions
def custom_accuracy(y_true, y_pred):
    return accuracy_score(y_true, y_pred)

def custom_balanced_accuracy(y_true, y_pred):
    return balanced_accuracy_score(y_true, y_pred)

def custom_auc_roc(y_true, y_probabilities):
    return roc_auc_score(y_true, y_probabilities[:,-1])

def custom_auc_pr(y_true, y_probabilities):
    return average_precision_score(y_true, y_probabilities[:,-1])
```

```python
def evaluate_model(model, x, y):
    acc_scores = cross_val_score(model, x, y, cv=skfold, scoring='accuracy')
    bal_acc_scores = cross_val_score(model, x, y, cv=skfold, scoring='balanced_accuracy')

    # Use cross_val_predict to get predicted probabilities for AUC-ROC and AUC-PR
    y_probabilities = cross_val_predict(model, x, y, cv=skfold, method='predict_proba')[:, -1]

    auc_roc_scores = cross_val_score(model, x, y, cv=skfold, scoring='roc_auc')
    auc_pr_scores = cross_val_score(model, x, y, cv=skfold, scoring='average_precision')

    return np.mean(acc_scores), np.mean(bal_acc_scores), np.mean(auc_roc_scores), np.mean(auc_pr_scores)
```
